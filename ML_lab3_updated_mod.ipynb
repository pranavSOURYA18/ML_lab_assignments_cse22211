```python
import pandas as pd
import numpy as np

# Load and preprocess the data
def load_and_preprocess_data(file_path):
    df = pd.read_csv(file_path)

    # Encode categorical columns using one-hot encoding
    df_encoded = pd.get_dummies(df, columns=['class', 'gender', 'race'])

    # Extract features and labels
    features = df_encoded.drop(columns=['y'])
    labels = df['y']

    return features, labels

# Get data for each class
def get_class_data(features, labels, class_value):
    return features[labels == class_value].to_numpy()

# Compute the centroid of a class
def compute_centroid(class_data):
    return class_data.mean(axis=0)

# Compute the intraclass spread
def compute_intraclass_spread(class_data, centroid):
    return np.mean(np.linalg.norm(class_data - centroid, axis=1))

# Compute the interclass distance between two centroids
def compute_interclass_distance(centroid_1, centroid_2):
    return np.linalg.norm(centroid_1 - centroid_2)

# Main function to load data and compute class characteristics
def main(file_path):
    features, labels = load_and_preprocess_data(file_path)

    print("Unique labels:", labels.unique())

    class_1_data = get_class_data(features, labels, 0)
    class_2_data = get_class_data(features, labels, 1)

    print("Shape of class_1_data (Class 0):", class_1_data.shape)
    print("Shape of class_2_data (Class 1):", class_2_data.shape)

    if class_1_data.size == 0 or class_2_data.size == 0:
        print("One of the classes is empty. Please check the class labels and data.")
        return

    # Convert to float
    class_1_data = class_1_data.astype(float)
    class_2_data = class_2_data.astype(float)

    # Compute centroids
    class_1_centroid = compute_centroid(class_1_data)
    class_2_centroid = compute_centroid(class_2_data)

    print("Class 1 Centroid (Class 0):", class_1_centroid)
    print("Class 2 Centroid (Class 1):", class_2_centroid)

    # Compute intraclass spreads
    class_1_spread = compute_intraclass_spread(class_1_data, class_1_centroid)
    class_2_spread = compute_intraclass_spread(class_2_data, class_2_centroid)

    print("Class 1 Intraclass Spread (Class 0):", class_1_spread)
    print("Class 2 Intraclass Spread (Class 1):", class_2_spread)

    # Compute interclass distance
    interclass_distance = compute_interclass_distance(class_1_centroid, class_2_centroid)

    print("Interclass Distance between Class 0 and Class 1:", interclass_distance)

# Example usage
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
main(file_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data from a CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

# Extract and clean the feature data (remove NaNs)
def get_clean_feature_data(df, feature):
    data = df[feature].dropna()
    return data

# Compute mean and variance of the feature data
def compute_statistics(data):
    mean = np.mean(data)
    variance = np.var(data)
    return mean, variance

# Plot histogram of the feature data
def plot_histogram(data, feature):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins='auto', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# Main function to run the process
def main(file_path, feature):
    df = load_data(file_path)

    # Extract and clean data for the specified feature
    data = get_clean_feature_data(df, feature)

    # Compute statistics
    mean, variance = compute_statistics(data)

    # Plot histogram
    plot_histogram(data, feature)

    # Print the mean and variance
    print(f"Mean of {feature}: {mean}")
    print(f"Variance of {feature}: {variance}")

# Example usage
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
feature = 'GPA'
main(file_path, feature)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from the CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

# Extract feature vectors from the dataset
def extract_vectors(df, index_1, index_2, feature_list):
    vector_1 = np.array(df.loc[index_1, feature_list])
    vector_2 = np.array(df.loc[index_2, feature_list])
    return vector_1, vector_2

# Compute Minkowski distance between two vectors
def minkowski_distance(x, y, r):
    return np.sum(np.abs(x - y) ** r) ** (1 / r)

# Calculate Minkowski distances for different r values
def calculate_distances(vector_1, vector_2, r_values):
    return [minkowski_distance(vector_1, vector_2, r) for r in r_values]

# Plot the Minkowski distances
def plot_distances(r_values, distances):
    plt.figure(figsize=(10, 6))
    plt.plot(r_values, distances, marker='o')
    plt.title('Minkowski Distance between Two Feature Vectors')
    plt.xlabel('r')
    plt.ylabel('Distance')
    plt.grid(True)
    plt.xticks(r_values)
    plt.show()

# Main function to orchestrate the process
def main(file_path, index_1, index_2, feature_list):
    df = load_data(file_path)

    # Extract vectors for the given indices and feature list
    vector_1, vector_2 = extract_vectors(df, index_1, index_2, feature_list)

    # Define r values and calculate Minkowski distances
    r_values = np.arange(1, 11)
    distances = calculate_distances(vector_1, vector_2, r_values)

    # Plot the distances
    plot_distances(r_values, distances)

    # Print the distances
    for r, distance in zip(r_values, distances):
        print(f"Minkowski Distance with r={r}: {distance}")

# Example usage
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
index_1 = 0
index_2 = 1
feature_list = ['GPA', 'Algebra', 'Calculus1']

main(file_path, index_1, index_2, feature_list)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load data from the CSV file
def load_data(file_path):
    return pd.read_csv(file_path)

# Filter the dataset for specific class values
def filter_binary_classes(df, class_1, class_2):
    df_binary = df[df['y'].isin([class_1, class_2])]
    return df_binary

# Preprocess features and labels by converting to numeric and handling missing values
def preprocess_data(df, feature_columns):
    features = df[feature_columns].apply(pd.to_numeric, errors='coerce')
    df_cleaned = df.dropna()
    features_cleaned = df_cleaned[feature_columns]
    labels_cleaned = df_cleaned['y']
    return features_cleaned, labels_cleaned

# Split the data into training and testing sets
def split_data(features, labels, test_size=0.3, random_state=42):
    return train_test_split(features, labels, test_size=test_size, random_state=random_state)

# Main function to coordinate the process
def main(file_path, class_1, class_2, feature_columns):
    df = load_data(file_path)

    # Check unique values in the target column
    print("Unique values in 'y':", df['y'].unique())

    # Filter the dataset for binary classes
    df_binary = filter_binary_classes(df, class_1, class_2)
    print("Shape of filtered DataFrame:", df_binary.shape)

    # Preprocess features and labels
    features, labels = preprocess_data(df_binary, feature_columns)
    print("Features shape after dropping missing values:", features.shape)
    print("Labels shape after dropping missing values:", labels.shape)

    # Check if data is available after preprocessing
    if features.shape[0] == 0 or labels.shape[0] == 0:
        print("No data available to split. Check your filtering and preprocessing steps.")
    else:
        # Split the data
        X_train, X_test, y_train, y_test = split_data(features, labels)

        # Print training and testing set shapes
        print("Training set shape:", X_train.shape)
        print("Testing set shape:", X_test.shape)

# Example usage
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
class_1 = 0
class_2 = 1
feature_columns = ['class', 'gender', 'race']

main(file_path, class_1, class_2, feature_columns)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load the dataset
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
df = pd.read_csv(file_path)

# Choose two classes to filter
class_1 = 0  # Example class
class_2 = 1  # Example class

# Ensure the dataset contains only these two classes
df_binary = df[df['y'].isin([class_1, class_2])]

# Define features (X) and labels (y)
features = df_binary[['class', 'gender', 'race']]  # Replace with actual feature column names
labels = df_binary['y']  # Replace with actual label column name

# Convert categorical features to numeric if necessary
features = features.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df_binary = df_binary.dropna()
features = df_binary[['class', 'gender', 'race']]
labels = df_binary['y']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# List to store accuracy for different values of k
accuracy_list = []

# Vary k from 1 to 11
for k in range(1, 12):
    # Train the kNN classifier with current k
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X_train, y_train)

    # Predict on the test set
    y_pred = neigh.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_list.append(accuracy)

# Plotting the accuracy vs k
plt.figure(figsize=(10, 6))
plt.plot(range(1, 12), accuracy_list, marker='o')
plt.title('kNN Classifier Accuracy for different values of k')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.xticks(range(1, 12))
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = r"C:/Users/Admin/Documents/Students_data.csv"
df = pd.read_csv(file_path)

# Choose two classes to filter
class_1 = 0  # Example class
class_2 = 1  # Example class

# Ensure the dataset contains only these two classes
df_binary = df[df['y'].isin([class_1, class_2])]

# Define features (X) and labels (y)
features = df_binary[['class', 'gender', 'race']]  # Replace with actual feature column names
labels = df_binary['y']  # Replace with actual label column name

# Convert categorical features to numeric if necessary
features = features.apply(pd.to_numeric, errors='coerce')

# Drop rows with missing values
df_binary = df_binary.dropna()
features = df_binary[['class', 'gender', 'race']]
labels = df_binary['y']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Train the kNN classifier with k=3
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X_train, y_train)

# Predict on the training set
y_train_pred = neigh.predict(X_train)

# Predict on the test set
y_test_pred = neigh.predict(X_test)

# Confusion matrix for training set
cm_train = confusion_matrix(y_train, y_train_pred)
print("Confusion Matrix for Training Set:\n", cm_train)

# Confusion matrix for test set
cm_test = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix for Test Set:\n", cm_test)

# Calculate precision, recall, and F1 score for training set
precision_train = precision_score(y_train, y_train_pred)
recall_train = recall_score(y_train, y_train_pred)
f1_train = f1_score(y_train, y_train_pred)
accuracy_train = accuracy_score(y_train, y_train_pred)

# Calculate precision, recall, and F1 score for test set
precision_test = precision_score(y_test, y_test_pred)
recall_test = recall_score(y_test, y_test_pred)
f1_test = f1_score(y_test, y_test_pred)
accuracy_test = accuracy_score(y_test, y_test_pred)

# Print the metrics for training set
print("\nTraining Set Performance Metrics:")
print(f"Accuracy: {accuracy_train:.2f}")
print(f"Precision: {precision_train:.2f}")
print(f"Recall: {recall_train:.2f}")
print(f"F1 Score: {f1_train:.2f}")

# Print the metrics for test set
print("\nTest Set Performance Metrics:")
print(f"Accuracy: {accuracy_test:.2f}")
print(f"Precision: {precision_test:.2f}")
print(f"Recall: {recall_test:.2f}")
print(f"F1 Score: {f1_test:.2f}")

# Plotting the confusion matrix for the test set
plt.figure(figsize=(8, 6))
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Test Set')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
